## Life Satisfaction Prediction

```python
# import libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
```

### Load and Prepare Data

```python
datafile = "datasets/lifesat/"

# Download CSV from http://stats.oecd.org/index.aspx?DataSetCode=BLI
oecd = pd.read_csv(datafile +"oecd_bli_2015.csv", thousands=',')
oecd = oecd[oecd["INEQUALITY"]=="TOT"]
oecd = oecd.pivot(index="Country", columns="Indicator", values="Value")
oecd.head()
```

```python
oecd["Life satisfaction"].head()
```

```python
# GDP per capita
# Download data from http://goo.gl/j1MSKe (=> imf.org)

gdp = pd.read_csv(datafile+"gdp_per_capita.csv", thousands=',', delimiter='\t',
                             encoding='latin1', na_values="n/a")
gdp.rename(columns={"2015": "GDP per capita"}, inplace=True)
gdp.set_index("Country", inplace=True)

gdp.head()
```

```python
# full country stats

full_country = pd.merge(left= oecd, right= gdp, left_index=True, right_index=True)
full_country.sort_values(by="GDP per capita", inplace=True)
full_country
```

```python
full_country[["GDP per capita", "Life satisfaction"]].loc["Ireland"]
```

```python
remove_indices = [0, 1, 6, 8, 33, 34, 35]
keep_indices = list(set(range(36)) - set(remove_indices))

sample_data = full_country[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
missing_data = full_country[["GDP per capita", 'Life satisfaction']].iloc[remove_indices]
```

```python
sample_data.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(9,5))
plt.axis([0, 60000, 0, 10])
position_text = {
    "Hungary": (5000, 1),
    "Korea": (18000, 1.7),
    "France": (29000, 2.4),
    "Australia": (40000, 3.0),
    "United States": (52000, 3.8),
}
for country, pos_text in position_text.items():
    pos_data_x, pos_data_y = sample_data.loc[country]
    country = "U.S." if country == "United States" else country
    plt.annotate(country, xy=(pos_data_x, pos_data_y), xytext=pos_text,
            arrowprops=dict(facecolor='black', width=0.5, shrink=0.1, headwidth=5))
    plt.plot(pos_data_x, pos_data_y, "ro")
plt.show()
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
Xsample = np.c_[sample_data["GDP per capita"]]
Ysample = np.c_[sample_data["Life satisfaction"]]

model.fit(Xsample,Ysample)
t0, t1 = model.intercept_[0], model.coef_[0][0]
t0, t1
```

```python
sample_data.plot(kind='scatter', x="GDP per capita", y='Life satisfaction', figsize=(9,5))
plt.axis([0, 60000, 0, 10])
X=np.linspace(0, 60000, 1000)
plt.plot(X, t0 + t1*X, "b")
plt.text(5000, 3.1, r"$\theta_0 = 4.85$", fontsize=14, color="b")
plt.text(5000, 2.2, r"$\theta_1 = 4.91 \times 10^{-5}$", fontsize=14, color="b")

plt.show()
```

```python
# make prediction
cyprus_gdp_per_capita = gdp.loc["Cyprus"]["GDP per capita"]
print(cyprus_gdp_per_capita)

cyprus_predicted_life_satisfaction = model.predict(cyprus_gdp_per_capita)[0][0]
cyprus_predicted_life_satisfaction
```
